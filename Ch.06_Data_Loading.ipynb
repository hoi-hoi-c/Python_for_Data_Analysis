# 6.1.5 XML과 HTML : Crawling 크롤링

- lxml: 큰 파일 빠르게 처리, 인터페이스 제공
- lxml.html : HTML 처리
- lxml.objectify : XML 처리

- 순서: 데이터 url 확인 -> requests 모듈로 불러오기 -> lxml로 파싱

- 파싱? 데이터 타입을 알아서 구별해 주는 기능 (h1, h2, a, 등등)
    + 웹페이지에서 긁어오면 모두 문자열로 취급되기 때문에 파싱이 필요하다.

```python
import requests

from lxml.html import parse
from io import StringIO

text = requests.get('https://finance.naver.com/').text
head = requests.get('https://finance.naver.com/').headers
```

```python
parsed = parse(StringIO(text))
doc = parsed.getroot() #파싱한 객체를 getroot()를 통해 저장한다. 
```

###### 각 table = header + data

```python
tables = doc.findall('.//table')
calls = tables[1]
puts = tables[2]
```

```python
rows = calls.findall('.//tr')
rows
```

th: header

td: data

```python
def _unpack(row, kind ='td'):
    elts = row.findall('.//%s' % kind)
    return [val.text_content().strip().split('\n')[0] for val in elts]
```

```python
_unpack(rows[0], kind ='th')
_unpack(rows[2], kind = 'td')
```

###### TextParser 클래스: pandas에서 제공하는 자동 형변환 클래스

```python
from pandas.io.parsers import TextParser
```

```python
def parse_options_data(table):
    rows = table.findall('.//tr')
    header = _unpack(rows[0], kind ='th')
    data = [_unpack(r) for r in rows[1:]]
    return TextParser(data, names = header).get_chunk() # TextParser 를 사용하여 데이터를 넣고, header 를 지정하여 데이터 타입을 알아서 추측
```

```python
call_data = parse_options_data(calls)
call_data
```

```python
put_data = parse_options_data(puts)
put_data
```
