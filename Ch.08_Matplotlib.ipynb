# 8.1.2 색상, 마커, 선 스타일
- ax.plot(x, y, linestyle ='--', color = 'k', marker = 'o', drawstyle = 'steps-post', label='')
    
    + marker : 꼭짓점에 대한 강조 'o', '^', 'x'
    + drawstyle: 그래프 스타일
    + label : 그리려는 선의 이름/ 하나의 표에 여러 그래프가 들어갈 때 각각의 이름을 붙여 식별
    + plt.legend(loc='best'): 라벨 표시 위치 설정

```python
smpldata1 = np.random.randn(20)
smpldata2 = np.random.randn(20)

plt.plot(smpldata1, 'g--', marker = 'x', label = 'line1')
plt.plot(smpldata2, 'bo-', marker = 'o', label = 'line2')
plt.legend(loc='best')
```

# 8.1.3 눈금, 라벨, 범례
제목,축이름, 눈금, 눈금 이름 설정

- ax.set_xticks([]) : X축 눈금의 범위와 숫자 설정
- ax.set_xticks([]) : X축 눈금 이름 설정
- ax.xlabel('') : X축 이름 설정
- ax.set_title('') : 서브 플롯의 제목 설정

```python
datax = np.arange(6)
figure, ax = plt.subplots(2,2)

ax[0,0].plot(datax, 'go--') # color='g', marker ='o', linestyle = '--'
ax[0,0].set_xticks([i for i in range(0,7)])
ax[0,0].set_yticks([i for i in range(0,10,2)])
ax[0,0].set_xticklabels(list('abcdef'), color = 'b')


ax[0,0].set_xlabel('x label', color='r')
ax[0,0].set_title('My First graph', color='g')
figure.subplots_adjust(wspace= 0.3, hspace= 0.5)
```

```python
#label 추가하기
datak = np.random.randn(10)
ax[0,1].plot(datak, 'b--', label = 'line1')
ax[0,1].plot(datak, 'rx--', label = 'line2')
ax[0,1].legend(loc='best')

figure
```

# 도형 추가하기
1. 도형 객체 만들기
    + 객체 이름 = plt.Rectangle([,], 가로길이, 세로길이, color=, alpha=)
        - 맨 앞 []은 도형의 왼쪽 하단부 꼭짓점
        - alpha : 투명도
    + 객체 이름 = plt.CIrcle([,], 반지름, color = )
    + 객체 이름 = plt.Polygon([(,),(,),(,)]) : 각각의 점을 잇는 도형
2. 서브 플롯에 `ax.add_patch(도형객체)` 이용하여 추가

```python
rect = plt.Rectangle([0.5, 0.5], 0.2, 0.4, color ='r')
circ = plt.Circle([0.2,0.2], 0.1, color = 'b' )
Poly = plt.Polygon([(0.75, 0),(0.8,0.25),(0.5, 0.3),(0.75,0)], color='g')

ax[1,1].add_patch(rect)
ax[1,1].add_patch(circ)
ax[1,1].add_patch(Poly)

figure
```

# Pandas로 그래프 그리기
pandas 객체는 기본적으로 행과 열 이름 등 많은 정보로 담고 있어 matplotlib로 그래프를 그릴 때보다 코드를 짧게 줄일 수 있다.

index가 x축으로 해석된다. 

- use_index = index 값을 x축으로 사용할지 여부
- xticks : x축의 눈금을 직접 설정
- xlim : x축의 한계 범위 설정
- pd객체.plot() : 기본값은 선으로 그려준다.

```python
s = pd.Series(np.random.randn(50).cumsum())
s.plot(use_index = False)
```

# DataFrame으로 plot 그리기
각 열의 header 이름 값이 범례가 된다. (선 하나씩)

dataframe에서만 사용하는 plot 옵션
- subplots : 각 column을 다른 서브플롯에 그릴지 여부
- sharex, sharey : 서브플롯끼리 축을 공유하고 눈금과 한계를 연결
- legend : 서브플롯의 범례 추가
- sort_columns : column을 알파벳 순서로 그린다.

```python
df = pd.DataFrame(np.random.randn(10,4).cumsum(0),
                 columns = [list('ABCD')],
                 index = np.arange(0,100,10))
df
```

```python
df.plot(xticks = [i for i in range(0,100,5)], grid = True, title = 'My graph by pandas')
```

```python
#각 선을 분리하여 서브플롯에 저장하기
df.plot(kind='line', grid = True, xticks=[i for i in range(0,100,5)],
       title = 'My graph by pandas', figsize = (10,7),
       subplots = True)
```

```python
df2 = pd.DataFrame(np.random.rand(10,5).cumsum(0),
                   columns = pd.Index(list('ABCDE')))
df2.index.name = "idx names"
df2.columns.name = "col names"
df2.plot()
df2
```

# 8.2.2 막대 그래프
- .plot(kind = 'bar' or 'barh') : bar 세로막대, barh 가로막대
- kde(kernel density estimate): 근사한 값을 추정하여 그래프를 그린다.

```python
#1차원 데이터
np.random.seed(11)
fig3, axes = plt.subplots(2,1)
data = pd.Series(np.random.randn(16), index=list('abcdefghijklmnop'))

data.plot(kind='bar', ax=axes[0], color = 'k', alpha = 0.3)
data.plot(kind='barh', ax=axes[1], color = 'r', alpha = 0.6)
data
```

# 8.2.3 히스토그램과 밀도 그래프

```python
comp1 = np.random.normal(0.1, size =200)
comp2 = np.random.normal(10,2, size =200)

values = pd.Series(np.concatenate([comp1, comp2]))

values.hist(bins=100, alpha =0.4, color ='k', normed = True)
values.plot(kind = 'kde', style = 'r--')
```
